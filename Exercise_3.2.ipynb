```python
import numpy as np
```

```python
x = np.random.uniform(low=-1, high=1, size=10000)
y = np.random.uniform(low=-1, high=1, size=10000)

darts = np.column_stack((x, y))
```

```python
in_circle_2d = 0
radius_2d = []

for dart in darts:
    radius = np.linalg.norm(dart)
    radius_2d.append(radius)

    if radius < 1:
        in_circle_2d += 1
```

```python
in_circle_2d / 10000
```

The area of the square dartboard of length 2 is 4, meanwhile the area of the circle of radius 1 is pi (approximately 3.1415). So the circle takes up approximately 78% of the circle. The uniform distribution leads to approximately 78.58% of the darts to land in the circle, which is very close to the percentage of the area the circle takes. So the number of darts the lands in the circle lines up with reasonable predictions.

```python
in_circle_list = [in_circle_2d]
radius_list = np.array(radius_2d)

for dim in (3, 4, 5, 6, 7, 8, 9, 10):
    z = np.random.uniform(low=-1, high=1, size=10000)
    darts = np.column_stack((darts, z))
    in_circle = 0
    radius_multi = []

    for dart in darts:
        radius = np.linalg.norm(dart)
        radius_multi.append(radius)

        if radius < 1:
            in_circle += 1

    radius_list = np.column_stack((radius_list, radius_multi))
    in_circle_list.append(in_circle)
```

```python
for i in range(9):
    print("Fraction that lands in circle in " + str(i + 1) + " dimensions:\t" + str(in_circle_list[i] / 10000))
```

```python
import matplotlib.pyplot as plt
plot_points = []
for i in in_circle_list:
    plot_points.append(i / 10000)

x = [2, 3, 4, 5, 6, 7, 8, 9, 10]
plt.title("Fraction of Darts landing in circle vs N-dimensions")
plt.plot(x, plot_points)
plt.show()
```

```python
for i in range(9):
    plt.title("Dart radius distribution at " + str(i + 2) + " dimensions")
    plt.boxplot(radius_list[:,i])
    plt.show()
```

We see that as N increases, the radius of each dart relative to the origin increases as well, so less and less darts end up landing in the N-dimensional circle.
