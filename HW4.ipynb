```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.random.standard_normal(size=(100, 100))
U, S, VT = np.linalg.svd(mat,full_matrices=False)

S = np.diag(S)

plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```

```python
def distribution(r, mean):
  if mean is True:
    return np.mean(S[:r])
  else:
    return np.median(S[:r])
```

```python
mat = np.random.standard_normal(size=(100, 100))
U, S, VT = np.linalg.svd(mat,full_matrices=False)

means = []
medians = []

for x in range(100):
  means.append(distribution(x, True))
  medians.append(distribution(x, False))
```

```python
plt.plot(means)
plt.show()
```

```python
plt.plot(medians)
plt.show()
```

```python
for x in (50, 200, 500, 1000):
  mat = np.random.standard_normal(size=(x, x))
  U, S, VT = np.linalg.svd(mat,full_matrices=False)

  means = []
  medians = []

  for i in range(x):
    means.append(distribution(i, True))
    medians.append(distribution(i, False))

  plt.figure(1)
  plt.semilogy(means)
  plt.title('Means: ' + str(x))
  plt.show()

  plt.figure(2)
  plt.plot(medians)
  plt.title('Medians: ' + str(x))
  plt.show()
```
